# Cálculos nas redes neurais (seção 3)


A primeira camada de uma rede é formada por neurônios que representam e descrevem a entrada do problema, onde cada neurônio presente está associado a um peso, conectando todos dessa camada a um próximo neurônio. Na chegada, todos os valores de entrada são multiplicados por seus respectivos pesos e tudo isso é somado. Ao fim, o valor da soma é passado pela função de ativação, que irá produzir um output para o determinado neurônio.


Aplicações de redes neurais se preocupam em achar os pesos mais adequados para cada neurônio de forma a aumentar a eficácia do modelo.


O modelo mais simples é o perceptron de uma camada. Seu uso é mais comum em problemas linearmente separáveis, o que não representa a maioria dos problemas reais. Para isso, existem redes neurais com um número maior de camadas.


No vídeo é mostrado o exemplo de um modelo que deve aprender a predizer valores do operador lógico xor, onde na primeira vez que o modelo tenta encontrar uma resposta, todos os seus pesos são colocados de forma aleatória, levando ele a performar mal.


Para medir o quão bem uma rede aprende, uma das formas é:


erro = resposta correta - resposta calculada


# Gradiente


Visto a representação do erro como uma função que é alterada conforme se muda os pesos da rede, ou seja, recebe os valores dos pesos e associa a eles um erro, a imagem dessa função pode ser conceitualmente entendida como um campo de vales e montanhas, os nos vales estão presentes aqueles erros menores, portando, também os melhores pesos.


O Gradiente procura caminhos na imagem dessa função de forma a encontrar o menor ponto possível.


Então, de forma resumida, o gradiente:


- Busca encontrar a melhor combinação de pesos
- Mostra o quanto ajustar os pesos


## Delta


O cálculo do gradiente é feito através da derivada da função de ativação, onde primeiro se é calculado ele usando o erro medido da primeira interação do modelo. Isso mostra qual passo o modelo deve tomar para melhorar.


Ao fim, para cada entrada diferente, é considerado a derivada da função de ativação do neurônio, algo que eu entendo como o quão o neurônio está propenso a mudar vezes o próprio peso de saída, isso multiplicado pelo Delta de saída que diz para onde o modelo deve caminhar para diminuir o erro.


## Back propagation


Dessa forma, os novos valores de cada peso são calculados de trás para frente nas camadas, sendo então o Delta saída o cálculo da direção de onde o modelo deve andar, e esse valor influencia toda a rede conforme novos pesos são calculados


# Bias


São valores extras colocados tanto na entrada como na saída, mudando o resultado com base na unidade de bias.


# MSE e RMSE


A forma acima de calcular o erro não é a mais adequada para modelos reais, para isso existem a Mean Square Error (MSR) e a Root Mean Square Error (RMSE).


## MSE


Calculado assim com o erro anterior, porém, é elevado ao quadrado o antigo resultado, o qual também é posteriormente somado e tirado a média.


O ato de elevar ao quadrado se traduz na penalização maior do erro para o modelo.


## RMSE


Idêntico ao MSE, porém, ao fim, é tirado a raíz da média calculada.


# Gradiente estocástico


Diferente do aprendizado anterior que recalculava os pesos de cada entrada, o gradiente estocástico irá atualizar os pesos com base em blocos da entrada do dataset, calculando o erro geral para cada bloco e atualizando o seu peso.


Essa abordagem ajuda a prevenir os mínimos locais e é mais rápido por não precisar carregar todos os dados na memória.


# Mini batch gradient descent


De forma a misturar os conceitos de ambos gradientes apresentados, ele irá escolher um número de registro para rodar e atualizar os pesos.


# Parâmetros


Learning rate = taxa de aprendizagem

Batch size = tamanho do lote

Epochs = épocas


# Funções de ativação


## Step function


Retorna apenas o valor 0 ou 1, e por isso, é usado em problemas com soluções lineares. Não usada em problemas mais complexos.


## Sigmoide


Retorna valores entre 0 e 1. Muito utilizada em retorno de probabilidade e problemas binários (de duas classes).


## Hyperbolic tangent


Retorna valores entre -1 e 1. Também usada para classificação binária e permite a maior consideração de números negativos.


## ReLU


Retorna 0 ou valores maiores que 0, não possuindo um valor máximo.


## Linear


Apenas retorna o valor de entrada. Usada em problemas de regressão.


## Softmax


Retorna probabilidade para problemas maiores que duas classes.

Um dos casos onde é possível usar as redes é no caso de classificação binária em câncer de mama, predizendo se há um câncer ou não.

## Keras


No curso é usada a biblioteca Keras, a qual proporciona a criação de modelos de deep learning com poucas linhas. Assuntos referente a ela são consultados em sua documentação.


Em primeiro lugar, o modelo Sequential é importado de keras.models, fazendo com que assim, ao criar esse modelo, deva ser adicionado cada camada.


Em específico para a criação da camada, a mesma é feita utilizando um objeto Dense da parte de layers do keras, onde para a criação desse objeto é fornecido dados como a quantidade de neurônios, a função de ativação, a inicialização dos pesos e a dimensão da entrada.


OBS: A camada Dense conecta todos os neurônios de uma camada com todos os neurônios da camada subsequente.


A última camada deve coincidir contextualmente com o problema dado.


Após a criação da rede, ela é compilada usando parâmetros como otimizador, função de perda e métrica.


Ao fim, a rede é treinada usando o método fit no classificador.


## Previsões e avaliação da rede neural


Com uma rede treinada, o método predict irá levar em consideração os dados de testes e irá predizê-los, para depois poder ser avaliado.


No curso, os métodos usados são a matriz de confusão e accuracy_score, ambas da sklearn.metrics.


O classificador também pode ser avaliado pelo método evaluate.


## Criando um otimizador


Os otimizadores do keras são inicializados como um objeto, possuindo um learning rate, decay (de quanto em quanto decrementa o learning rate) e clipvalue (limita tamanho do passo do gradiente)


## Validação cruzada


Assim como estudado anteriormente neste bootcamp, a mesma técnica que validação cruzada pode ser usada na rede neural.


A validação é feita usando o KerasClassifier junto com cross_val_score.


## Dropout


Como uma forma de prevenção ao overfitting, o dropout irá zerar alguns neurônios da camada de entrada, ou seja, parte da entrada é desconsiderada.


O dropout é implementado após a adição de alguma camada, fazendo com que o efeito seja feito nela.


O dropout por padrão deve ser considerado ao implementar uma rede neural, visto que o alto número de parâmetros na criação de uma rede pode levá-la facilmente ao overfitting.


## Tuning


Ao contrário do processo feito até agora de passagem de parâmetros únicos para uma rede, o método de GridSearch irá receber possíveis parâmetros que o modelo pode receber, e a partir disso, fazer diferentes combinações que possuirão diferentes acurácias a fim de encontrar aquele modelo que melhor performou, encontrando assim bons parâmetros.


## Modelo em produção


Uma vez o modelo treinado e tunado, ele pode ser salvo em um arquivo json para que o mesmo modelo possa ser usado futuramente ou por outra pessoa.


OBS: Os pesos são salvos a parte em um arquivo .h5


# Classificação multiclasse iris (seção 5)


A base de dados íris consiste em determinar qual flor está sendo analisada a partir de atributos como tamanho da pétala, da sépala e entre outros.


Para esse problema, existem 3 classes de flores possíveis, tornando assim um problema de multiclasse.


Essas três classes se traduzem em adicionar na última camada da rede apenas 3 neurônios, cada um representando uma classificação. OBS: A função de ativação a ser usada deve ser a softmax


Para a compilação da rede, a função de loss e a métrica são alteradas para satisfazerem o contexto de multiclasse.


Para a rede conseguir compreender os dados de output do dataset, deve-se garantir que os mesmos sejam numéricos, aplicando transformações caso necessário, em casos como o do curso que vieram como strings. (LabelEnconder é usado no curso para resolver esse problema)


Além disso, o formato dos dados devem ter a quantidade de dimensões igual ao número de neurônios da última camada, para que a rede consiga comparar resultados. (np.utils de keras.utils é usado para isso)


Os formatos são desfeitos para que possa ser analisado o modelo usando a matriz de confusão.


O processo de validação cruzada também é aplicada, mas dessa vez é usado a média e o desvio padrão dos resultados para uma melhor análise.


# Base de carros usados (seção 6)


O objetivo é determinar qual o preço do veículo a partir de atributos que o descrevem. Diferente dos problemas antes visto, que eram de classificação, esse problema exige um valor real como resposta, sendo assim um problema de regressão.


Em primeiro lugar, é analisado o dataset e excluídas aquelas colunas que não possuem relação com o valor final do carro.


O value counts também é usado para extrair informações sobre as colunas, e uma vez encontradas aquelas que possuem um desbalanço de dados no value counts para aquele atributo, são excluídas.


O mesmo processo de LabelEncoder é usado para tratar dos dados não numéricos.


## one hot encoder


Onde antes havia dados categóricos do LabelEncoder, ou seja, uma única coluna possuindo valores como 1,2,3 e etc. Agora será criado um número de colunas igual a quantidade de classes possíveis desse atributo. Assim, para uma classe, apenas uma coluna correspondente a ela irá receber o valor 1, enquanto outras colunas recebem 0


## Criação rede neural


A rede é criada com a função de loss e métrica como mean_absolute_error visto que se trata de um problema de regressão.


A última camada da rede possui apenas um neurônio, esse obtendo a função de ativação linear.
