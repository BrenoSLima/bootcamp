# Redes neurais convolucionais

Difere-se de uma rede neural tradicional apenas na camada de entrada, a qual sofre uma transformação nos dados. Em geral usada em visão computacional, trazendo uma melhor performance melhor do até então outro melhor modelo da área, a SVM. Ao decorrer do processo de aprendizagem, extrai características mais importantes de um problema.


# Operador de convolução


Convolução é o processo de alterar cada elemento na imagem com base em seus vizinhos, ponderado por um kernel. Ou seja, a imagem original será escaneada por uma matriz pequena, varrendo cada pixel e alterando ele com base nos pixels em sua volta.


A depender dos valores dessa matriz, pode-se atingir diferentes objetivos, como detectar bordas, borrar a imagem e entre outros.


O Kernel é o processo de alterar a imagem original através de uma matriz específica, onde a palavra kernel também é usada para se referir a matriz. Porém, no contexto de redes neurais convolucionais, essa matriz recebe o nome de detector de características (feature detector).


Levando em consideração que ao analisar um pixel da imagem original, os seus vizinhos também estarão sendo considerados, sendo assim, pelo processo de convolução, se ao fim da conta o pixel for atualizado para um valor alto, significa que aquela área analisada pelo kernel possui características importantes.


Após a convolução, a resolução da imagem (feature map) é diminuída, facilitando assim seu processamento.


Ao fim da convolução, a função ReLU é aplicada em cada pixel, fazendo com que pixels negativos passem a ser 0.


Outra peculiaridade dessa rede neural é ela ter uma camada de convolução, onde diferentes detectores de características são testados e escolhidos. Ou seja, a própria rede escolhe qual feature detector (qual transformação) será o melhor para o problema.


# Pooling


Serve para enfatizar ainda mais as características do problema, onde o feature map será reduzido ainda mais para uma imagem menor. Esse processo é feito também escaneando o feature map, e agora para cada N pixels analisado (no exemplo do curso 4 pixels), escolhe o pixel com maior valor para compor a nova imagem.


Reduz o overfitting e ruídos desnecessários.


## Max polling


Usado em redes neurais convolucionais para focar nas características mais relevantes, extraindo o maior valor na região do pooling.


# Flattening


Processo de transformar uma matriz em um vetor, para que seja possível o vetor ser alimentado na camada de entrada da rede neural.


# Rede neural densa


É uma rede assim como as outras vistas no curso, possuindo treinamento com a descida do gradiente, onde uma vez obtendo o erro, volta-se até os processos iniciais, escolhendo novos features detector para o processo de novo.

# Estrutura da rede neural


O mesmo processo teórico explicado é seguido, e a implementação não se difere muito das outras redes neurais. A rede convolucional também irá seguir os padrões do Keras, onde a partir de um classificador, cada rede pode ser adicionada em ordem.

# Augmentation


Aumento da quantidade de imagens (augmentation), é o processo de criar novas imagens a partir do conjunto inicial de dados. Isso é feito mudando alguns pixels de imagens, rotacionando elas e etc.
