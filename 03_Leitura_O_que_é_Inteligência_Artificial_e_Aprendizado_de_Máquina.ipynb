# O que é inteligência artificial?


O objetivo da IA é permitir que máquinas observem, pensem e ajam como humanos, onde, para tal, é utilizado o machine learning como ferramenta, fazendo com que máquinas aprendam de acordo com dados apresentados.


# O que é aprendizado de máquina?


Como dito anteriormente, o aprendizado de máquina é o ato das máquinas aprenderem por si só, sem serem programadas. Isso é feito a partir de um conjunto de dados, o qual será usado para a máquina aprender padrões e características, criando regras que os aproximam, e a partir disso, conseguir aplicar essas regras para o mundo real.

## Como as máquinas aprendem

Primeiro, os dados são divididos entre conjuntos de treino e de validação.

### Conjunto de treino

O modelo não treinado é alimentado com os dados de treino para que consiga interpretar os padrões e regras nos dados. Os modelos aprendem por meio da indução, um processo de raciocínio que generaliza essas informações de forma a obter uma função/fórmula que descreve a relação entre dois ou mais elementos no conjunto de dados.

### Conjunto de validação

São conjuntos de dados desconhecidos pelo modelo, onde testam a função/fórmula gerada. Esses dados fornecem uma avaliação imparcial de quão bem o modelo se ajusta a dados não presentes no conjunto de treino.

# Os tipos de aprendizado de máquina

## Aprendizado Supervisionado

Nesse aprendizado, os dados são rotulados pelo ser humano, e o objetivo da máquina é criar uma função de mapeamento (f), que a partir dos dados rotulados (x), consiga gerar dados de saída (y') com base na função para novos dados desconhecidos (x').

Um exemplo disso é a regressão linear, onde a partir dos dados rotulados, o modelo irá aproximar a função de uma reta que melhor se encaixe nesses pontos, para então gerar novos dados poderem ser respondidos pelo modelo.

A partir desse princípio, pode-se criar um modelo também linear de classificação, onde sua função será classificar os dados, como se um email é spam ou não. A reta servirá para distinguir classes diferentes nos dados.

## Aprendizado não Supervisionado

Já para o aprendizado não supervisionado, os dados não são rotulados, e o objetivo desse modelo é atuar nesses dados sem orientação, a fim de encontrar agrupamentos e padrões naturais dentro dos dados. Um exemplo desses casos são os modelos de clusterização.

## Aprendizado por Reforço

Esses modelos aprendem interagindo com o seu ambiente, assim como nós humanos e outros animais, trazendo assim informações sobre causa e efeito. O objetivo de um agente de Aprendizado por Reforço é coletar o maior número possível de recompensas, as quais são obtidas através de um processo repetitivo de tentativa e erro.
