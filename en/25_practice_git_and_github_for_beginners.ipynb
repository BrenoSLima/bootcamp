# Git and Github for beginners


## I understand what git and github are


Git is a version control system that aims to organize different phases of a project. In other words, it stores past modifications of each file, allowing the user to access any version of it.


Unlike other ways of versioning code, git, instead of just saving the changes to each file, stores a snapshot of them all each time a new version is created.


While git is the locally applied version control system, github is the place where you can store code and projects.


## Configuring initial git

## Git essentials

When creating a folder for the project, the command

```python
git init
```

It should be used to properly initialize git, thus preparing the folder to start the project.

### File status lifecycle

For each file, there may be the following statuses:

* untracked: file has just been added and is not recognized by git.

* unmodified: file recognized by git, but not modified.

* modified: file recognized by git and modified.

* staged: files ready to be updated to new version.

Several commands with different functionalities are used below, which are listed below.

Show the status of each file:

```python
git status
```

Add file to staged state:

```python
git add file.txt
```

If a file already added is modified, it is considered modified, requiring the addition to be made again.

Create a repository version:

```python
git commit -m "mensagem"
```

Information such as author, date, and identification number are saved for each commit. The message must be clear and convey what was done in that project update.

Display history of past commits and their information:

```python
git log
```

Filter log by author:

```python
git log --author="Breno"
```

Summary log, which only shows the author, how many and what their commits were:

```python
git shortlog
```

Show more information about a commit and its changes:

```python
git show hash_code
```

Show differences between current changes and the latest version of the project:

```python
git diff
```

Used as a form of review before committing.

List modified filenames:

```python
git diff --name-only
```

## Undoing changes

If a file has been modified but not added, to undo the changes to it, simply use:

```python
git checkout file.txt
```

If it has been added, the files can be taken out of staged mode with:

```python
git reset HEAD
```

And then the checkout can be done.

If you need to undo a commit, there are different ways to do it:

--soft: undo commit leaving files as staged, ready to be committed again

--mixed: undo commit leaving files as modified.

--hard: undo all changes made to the commit.

```python
git reset --soft hash_code
```

## Remote repositories

The remote repository used in the course is on github, where it is created by the website interface. At the end of creation, the command that must be used locally to connect the repositories is displayed on the website.

```python
git remote add origin git@github.com:brenolima/github-course.git
```

To send all information from the local repository to the remote one:

```python
git push -u origin master
```

If you need to copy a repository that is already created to your machine:

```python
git clone git@github.com:brenolima/...
```

To change another user's repository, through github you must fork the other repository, causing a copy to be made for you, and once your changes are made, you can request a pull request for the original repository.

## Branching


So far in the course you have learned that there is a sequence of commits that make up the project, however, the concept of branches can be applied to branch the repository. Instead of having just one project update sequence, it has different branches. If the project reaches a level where two programmers need to work on the project, and each program a feature, each can create a branch and use separate workspaces.

Create branch

```python
git checkout -b branch_name
```

List branches

```python
git branch
```

Go to branch

```python
git checkout branch_name
```

Delete branch

```python
git branch -D branch_name
```

## Merge

Merge is an integration approach that combines two lines of development, as in the case of two developers working on functionality (branches) of the same application and need to join them. When you create a new branch to work on a specific functionality, for example, and want to incorporate these changes back into the main branch, merge is used.

Pros:

* Non-destructive operation

Cons:

* Extra commit

*Polluted history

```python
git merge test
```

Based on the current branch, it is merged into the other branch in command.

## Rebase

Rebase, on the other hand, is an integration technique that rewrites development history, adjusting the current branch's commits to be based directly on the target branch's latest commit.

Pros:

* Avoid extra commits

* Linear history

Cons:

* Loses chronological order

```python
git rebase test
```
