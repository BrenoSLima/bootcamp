# Data types


Different data require different types of tools, for this it is necessary to know how to identify and deal with these cases, which may include data of the following types: Numeric, categorical and ordinal.


## Numerics


They represent quantitative values, that is, they can be measured. They are discrete or continuous.


## Categorical


As they do not have mathematical interpretations, they are considered qualitative data, which describe cases such as: nationality, skin color, state of residence, etc.


## Ordinals


Mixture of numerical and categorical data, for example a voting system of up to 5 stars, where 3 stars is arguably better than 1 star, and can also be interpreted mathematically.


# Average


Sum of all values divided by the number of existing values.


# Median


Sorting data and selecting what is present in the middle of the sort. It is also a good alternative to the mean, as the median is less sensitive to outliers.


In the course example, in an income dataset, Jeff Bezos was added with 1 million, which caused a large change in the average, but not in the median.


# Fashion


Value that is most repeated in the data, applied only to discrete data.


# Standard deviation


Reports the degree of dispersion of a data set.


# Variance


Average of the squared distances of the values in relation to the mean of the data.


The act of squaring provides more information when there are outliers, and also makes it possible to compare negative numbers with positive ones.


# Population and sample


While the population represents the data set that has all measured values, the sample is a subset of the population.


In the case of the sample, the average must be calculated taking into account the number of data present minus one in the divisor.


# probability density functions


For a range of numbers in a distribution, reports the probability of a random continuous number being present in this range.


Therefore, intervals closer to the mean, and therefore with smaller standard deviations, have greater probabilities.


# Probability mass function


Similar to the density function, it is a way of visualizing the probability of occurrence of data, but only for discrete data.


# Percentiles


Reports a point where there is a certain percentage of cases in the data that have a value less than the given point.


The tenth percentile of a data set is a value where values below it represent 10% of the data.


# Time


The first moment is the mean, while the second moment is the variance.


The third moment is called "skew", representing in a way a "slanted normal distribution"


Fourth moment is Kurtosis, equivalent to how "thick" the trace is and how "pointy" the peak of a distribution is. Higher peaks have larger Kurtosis

# Covariance


It measures how two variables vary in a dataset, where data that is not very related to each other has low covariance. However, in the end it is difficult to say when it is low or high depending on the data measurement units.


# Correlation


Correlation has the same purpose as covariance, but both variables are divided by the standard deviation, meaning that:


-1 = Perfect inverse correlation
0 = No correlation
1 = Perfect correlation


# Conditional probability


Informs the probability of an event occurring, given that a previous event occurred.


P(B|A) = P(A, B) / P(A)


P(B|A) -> Probability of B occurring given that A has already occurred.


P(A, B) -> Joint probability of A and B occurring simultaneously.


P(A) -> Probability of event A occurring, independently of event B.


This is useful in scenarios where events are related and you need to understand how the occurrence of one event affects the probability of another event.


# Bayes' Theorem


P(A∣B)= (P(B) * P(B∣A)) / P(A)
​
P(A|B) -> Probability of event A occurring given that event B has already occurred


P(B|A) -> Probability of event B occurring given that event A has already occurred.


P(A) -> Prior probability that event A occurs


P(B) -> Probability of evidence B occurring, regardless of A.


In short, it helps you calculate the probability of A given B after considering the additional information provided by B.
