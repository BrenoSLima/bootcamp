# Convolutional neural networks

It differs from a traditional neural network only in the input layer, which undergoes a transformation in the data. Generally used in computer vision, bringing better performance than the other best model in the area, SVM. During the learning process, it extracts the most important characteristics of a problem.


# Convolution operator


Convolution is the process of changing each element in the image based on its neighbors, weighted by a kernel. In other words, the original image will be scanned by a small matrix, scanning each pixel and changing it based on the pixels around it.


Depending on the values of this matrix, different objectives can be achieved, such as detecting edges, blurring the image, among others.


Kernel is the process of changing the original image through a specific matrix, where the word kernel is also used to refer to the matrix. However, in the context of convolutional neural networks, this matrix is called a feature detector.


Taking into account that when analyzing a pixel of the original image, its neighbors will also be considered, therefore, through the convolution process, if at the end of the day the pixel is updated to a high value, it means that that area analyzed by the kernel has important features.


After convolution, the resolution of the image (feature map) is reduced, thus facilitating its processing.


At the end of the convolution, the ReLU function is applied to each pixel, causing negative pixels to become 0.


Another peculiarity of this neural network is that it has a convolution layer, where different feature detectors are tested and chosen. In other words, the network itself chooses which detector feature (which transformation) will be best for the problem.


# Pooling


It serves to further emphasize the characteristics of the problem, where the feature map will be further reduced to a smaller image. This process is also done by scanning the feature map, and now for each N pixels analyzed (in the course example 4 pixels), the pixel with the highest value is chosen to compose the new image.


Reduces overfitting and unnecessary noise.


## Max polling


Used in convolutional neural networks to focus on the most relevant features, extracting the highest value in the pooling region.


# Flattening


Process of transforming a matrix into a vector, so that the vector can be fed into the input layer of the neural network.


# Dense neural network


It is a network like the others seen in the course, with gradient descent training, where once the error is obtained, it goes back to the initial processes, choosing new detector features for the process again.
